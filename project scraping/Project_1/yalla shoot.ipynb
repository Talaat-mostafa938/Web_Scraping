```python
import requests
from bs4 import BeautifulSoup
import csv
import pandas as pd
```

```python
url = "https://shotz.yalla-shoot-tv.live/matches-today/"
```

```python
respons = requests.get(url).text
```

```python
respons
```

```python
html_content = BeautifulSoup(respons,"html")
```

```python
match = html_content.find("div",class_="match-container commingsoon")
```

```python
right_team = match.find("div" , class_="right-team").text.strip()
```

```python
right_team
```

```python
left_team = match.find("div" , class_="left-team").text.strip()
```

```python
left_team
```

```python
result_match = match.find("div" , class_="result")
```

```python
result_match.text
```

```python
time_start = match.find("div" , id="match-time")
```

```python
time_start.text
```

```python
csvfile = open("yalla.csv" , "w" , encoding='utf-8')
write = csv.writer(csvfile)
```

```python
write.writerow(["right team" , "result" , "left team" , "time"])
```

```python
matchs =  html_content.find_all("div",class_="match-container commingsoon")
```

```python
len(matchs)
```

```python
for match in matchs:
    right_team = match.find("div" , class_="right-team").text.strip()
    left_team = match.find("div" , class_="left-team").text.strip()
    result_match = match.find("div" , class_="result").text
    time_start = match.find("div" , id="match-time").text
    write.writerow([right_team, result_match , left_team , time_start ])
```

```python
csvfile.close()
```
